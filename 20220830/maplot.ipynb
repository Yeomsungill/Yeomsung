```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[1,4,2,8])
plt.show()
```

```python
x = [1,2,3,4]
y = [1,3,4,9]

plt.plot(x,y,"--r")
plt.plot(x, [4,2,6,5], "oy")
plt.xlabel("class")  # x 
plt.ylabel("count")  # y 
plt.title("class & count") #  
plt.legend(["A","B"], loc=0) #  
plt.show()
```

```python
# subplot
plt.subplot(1 , 2, 1)
plt.plot([1,2,3,4])
plt.subplot(1,2,2)
plt.plot([8,7,6,5])
plt.show()
```
