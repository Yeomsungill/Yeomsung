```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[1,4,2,8])
plt.show()
```

```python
x = [1,2,3,4]
y = [1,3,4,9]

plt.plot(x,y,"--r")
plt.plot(x, [4,2,6,5], "oy")
plt.xlabel("class")  # x축의 이름
plt.ylabel("count")  # y축의 이름
plt.title("class & count") # 그래프 제목
plt.legend(["A","B"], loc=0) # 범례 추가
plt.show()
```

```python
# subplot
plt.subplot(1 , 2, 1)
plt.plot([1,2,3,4])
plt.subplot(1,2,2)
plt.plot([8,7,6,5])
plt.show()
```

```python
# 막대 그래프
x = [1,2,3,4]
y = [10,2,5,6]

plt.bar(x,y)
plt.show()
```

```python
import numpy as np
```

```python
# 산점도 그래프
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y)
plt.show()
```

```python
# 산점도 그래프 옵션 추가(영역 확장,색상 변경, 투명도 조절)

color = np.random.rand(50)
area = (20 * np.random.rand(50))**2

plt.scatter(x,y, s=area, c=color, alpha=0.7)
plt.show()
```
