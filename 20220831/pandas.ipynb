 #### 파생변수 생성
 1. 원가, 단가, 판대단위 값을 기준으로 
 2. 총수익, 총비용, 총이윤 새로운 컬럼을 생성
 3. 총수익 = 단가 * 판대단위
 4. 총비용 = 원가 * 판매단위
 5. 총이윤 = 총수익 - 총비용

```python
# 새로운 컬럼을 추가 하는 방법
# 데이터 프레임명[새로운 컬럼의이름] = 컬럼에 들어갈 값(Series,list)
# 단가 컬럼 * 판매단위 컬럼

sales["총수익"] = sales["Unit Price"] * sales["Units Sold"]
sales.head()
```

```python
# 총비용, 총이윤 생성
sales["총비용"] = sales["Unit Cost"] * sales["Units Sold"]
sales["총이윤"] = sales["총수익"] - sales["총비용"]
sales.head()
```

###  shift()
1. shift(n) -> n을 공란으로 두게 되면 기본값이 1
2. n 값만큼 인덱스를 이동하여 그 행의 값을 가지게 된다.


### diff()
1. 한 객체 내에서 열과 열 / 행과 행 차이를 출력

```python
# shift() 함수를 이용하는 방법
# 몇개의 행을 이동 할 것인지 아무것도 적지 않으면 바로 위의 인덱스 설정
a_copy["확진자"] - a_copy["확진자"].shift()
```

```python
## diff()
a_copy["확진자"].diff()
```

```python
## 일일확진자 , 일일사망자라는 컬럼을 추가
a_copy["일일확진자"]= a_copy["확진자"] - a_copy["확진자"].shift()
a_copy["일일사망자"]= a_copy["사망자"].diff()
a_copy
```
